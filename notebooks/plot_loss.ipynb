```python
# =========================================
# PLOT DE CURVAS DE ENTRENAMIENTO
# =========================================

import pandas as pd
import matplotlib.pyplot as plt
import os

# Crear carpeta para plots si no existe
os.makedirs("../outputs/plots", exist_ok=True)

# Cargar archivo CSV generado durante el entrenamiento
loss_csv_path = "../outputs/logs/loss_history.csv"

df = pd.read_csv(loss_csv_path)

# Mostrar primeras filas
df.head()
```

```python

# =========================================
# CURVAS DE ENTRENAMIENTO + ANLISIS
# =========================================

best_epoch = df.loc[df["val_loss"].idxmin(), "epoch"]
best_val_loss = df["val_loss"].min()

plt.figure(figsize=(10, 6))

plt.plot(df["epoch"], df["train_loss"],
         marker="o", label="Train Loss")

plt.plot(df["epoch"], df["val_loss"],
         marker="o", label="Validation Loss")

# Marcar mejor epoch
plt.axvline(x=best_epoch, color="green",
            linestyle="--", alpha=0.7,
            label=f"Best Epoch = {best_epoch}")

# Anotacin del mnimo
plt.scatter(best_epoch, best_val_loss, color="green")
plt.text(best_epoch, best_val_loss + 0.005,
         f"Min Val Loss = {best_val_loss:.3f}",
         ha="center", fontsize=10)

plt.title("Curvas de Entrenamiento  Train vs Validation Loss", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()

# Guardar grfico
save_path = "../outputs/plots/loss_curve_analysis.png"
plt.savefig(save_path, dpi=140)
plt.show()

print(f" Grfico con anlisis guardado en: {save_path}")

```

La imagen muestra las curvas de prdida de entrenamiento y validacin a lo largo
de las primeras pocas. Se observa una disminucin progresiva de ambas curvas, las
cuales se mantienen cercanas entre s durante todo el entrenamiento. Este
comportamiento indica un aprendizaje estable del modelo y ausencia de fenmenos de
overfitting o underfitting en las etapas iniciales del entrenamiento.
